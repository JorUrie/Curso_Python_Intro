```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.loadtxt('x1')
y = np.loadtxt('y1')
```

```python
plt.plot(x, y, '-.',color = 'darkgoldenrod')
plt.grid()
plt.title('Función seno')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.savefig('Graficas_1', bbox_inches = 'tight')
plt.show()
```
