```python
# Cargando bibliotecas
import pandas as pd 
import datetime as dt
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
# Solo se cargan los datos
Year = pd.read_csv('Data-MS.csv', header = 0, usecols = ['Year'])
Month = pd.read_csv('Data-MS.csv', header = 0, usecols = ['Month'])
MS = pd.read_csv('Data-MS.csv', header = 0, usecols = ['MS'])
#print(Year, Month, MS)
print(Year)
```

```python
# Se tabulan los datos cargados
Jahr = pd.DataFrame(Year)
Monat = pd.DataFrame(Month)
SS = pd.DataFrame(MS)
print(Jahr, type(Jahr))
```

```python
# Pasando la informacion con un determinado formato a lista
Anne = Jahr.values.tolist()
Mois = Monat.values.tolist()
SP = SS.values.tolist()
#print(Mois)
```

```python
# Una lista de varias lista en solo una lista 
A = []
M = []
S = []
for i in range(len(Jahr)):
    A.append(Anne[i][0])
    M.append(Mois[i][0])
    S.append(SP[i][0])
#print(S)
```

```python
# Juntando agno y el mes
Date = []
for j in range(len(S)):
    Date.extend([str(M[j]) + '/' + str(A[j])])
print(type(Date))
```

```python
'''Se pasa de una lista a un arreglo'''
Date = np.array(Date)
SS = np.array(S)
#print(Date)
#print('''Esto es un curso de python

#Se llama introduccion a python

#Me gusta progrmar en Python''')
```

```python
# Transformando cada elemento de la lista en una fecha
chars = '[, ]'
DateSS = []
i = 0 
for i in range(len(Date)):
    DateSS.append(Date[i].translate(str.maketrans('', '', chars)))
#print(DateSS)
# Convirtiendo a formato de fecha
x = [dt.datetime.strptime(d, '%m/%Y').date() for d in DateSS]
#print(x)
```

```python
# Hiendo la grfica de la informacion cargada
plt.rcParams['axes.facecolor'] = 'white' # Cambia el color del fondo de la grafica
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%m/%Y')) # Espera un formato de fecha
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval = 4080)) # Imprime diferentes elementos de la fecha
plt.plot(x, SS, '.', linewidth = 1, color = 'indigo')
plt.gcf().autofmt_xdate()
plt.title('Sunspots', fontsize = 18)
plt.xlabel('Date', fontsize = 12)
plt.ylabel('Number of Susnspots', fontsize = 12)
plt.grid()
plt.savefig('Suspots', bbox_inches = 'tight')
plt.plot()
```
