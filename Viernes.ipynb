```python
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
import pandas as pd
```

```python
''' Cargando datos de mi archivo csv '''
Lat = pd.read_csv('SSNMX_catalogo_19000601_20230630_CHIH.csv', header = 4, usecols = ['Latitud'])
Lon = pd.read_csv('SSNMX_catalogo_19000601_20230630_CHIH.csv', header = 4, usecols = ['Longitud'])
Deep = pd.read_csv('SSNMX_catalogo_19000601_20230630_CHIH.csv', header = 4, usecols = ['Profundidad'])
#print(Lat)
```

```python
''' Dandole formato antes de convertirlo a lista '''
lat = pd.DataFrame(Lat)
lon = pd.DataFrame(Lon)
deep = pd.DataFrame(Deep)
#print(lon)
```

```python
'''Transformando la informacion cargada en lista'''
d1 = deep.values.tolist()
la1 = lat.values.tolist()
lo1 = lon.values.tolist()
#print(d1, type(d1))
```

```python
''' Haciendo una mutilista en una lista '''
d = []
latitud = []
longitud = []
for i in range(len(d1)): 
    d.append(d1[i][0])
    latitud.append(la1[i][0])
    longitud.append(lo1[i][0])
#print(d, type(d))
```

```python
'''Quitar los nan que estaban contenidos en los ultimos 7 elementos de la lista de profundidad
Quitar los elementos que no se pueden convertir en float porque son de tipo string'''
D = []
Lat = []
Lon =[]
for j in range(len(d)-7):
    if d[j] == 'en revision':
        D.append(0)
        Lat.append(0)
        Lon.append(0)
    else:
        D.append(d[j])
        Lat.append(latitud[j])
        Lon.append(longitud[j])
print(len(Lon))
```

```python
'''Se empieza hacer la grafica en 3D '''
''' Se inicia haciendo una figura '''
fig = plt.figure(figsize = (16, 12))
''' Se indica que se quiere hacer una grafica en 3 dimensiones '''
axes = plt.axes(projection = '3d')
'''Se introduce la informacion que se va a graficar'''
axes.scatter3D(D, Lat, Lon, color = 'navy')
# Formato de la figura
axes.set_title('Localizacion de los eventos sismicos ocurridos en Oaxca', fontsize = 16, fontweight = 'bold')
axes.set_xlabel('Longitud [km]', fontsize = 14)
axes.set_ylabel('Latitud [km]', fontsize = 14)
axes.set_zlabel('Profundidad [km]', fontsize = 14)
plt.tight_layout()
plt.grid()
plt.savefig('Posicion_de_los_sismos_en_Oaxaca', bbox_inches = 'tight')
plt.show()
```
