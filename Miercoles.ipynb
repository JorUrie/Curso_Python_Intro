```python
'''Se define una funcion en python que nos permita trazar una funcion
n: Es el numero de elementos que debe tener la lista'''
def Vektor():
    n = int(input('Numero de muestras: '))
    x = []
    for i in range(n):
        x.append(i)
    return x, len(x)
```

```python
# Se llama las  graficas que se hicieron ayer
x = np.loadtxt('x1')
y = np.loadtxt('y1')
#print(x)
# Solo hacer la grafica para saber si vamos bien
plt.plot(x, y)
```

# Funcion Seno Cardinal
$\ f(x) = sin(x)/x = sinc(x) \$

# Funcion por partes
$\ f(x) = cos(x)*sin(x)^-8 \\ \\ \\ \\ \\ \\ x < 49  \$

$\ f(x) = sin(x)*cos(x)^2 \\ \\ \\ \\ \\ \\ x >= 49  \$

```python
# Se traza la funcion seno cardinal
SinCar = []
for j in range(Val2):
    SinCar.append(np.sin(j)/j)
plt.plot(x, SinCar)
```

```python
'''Multiplicacion de dos vectores
El que ya teniaos hecho y el seno cardinal'''
Neue = y*SinCar
#print(Neue)
plt.plot(x, Neue)
```


# Funcion por partes
$\ f(x) = cos(x)*sin(x)^{-8} \\ \\ \\ \\ \\ \\ x < 49  \$

$\ f(x) = sin(x)*cos(x)^{1/2} \\ \\ \\ \\ \\ \\ x >= 49  \$

```python
'''Se traza la funcion parametrica
Con las funciones que multiplicamos, sus valores son introducidos dentro de la funcion parametrica'''
funPar = []
for i in range(Val2):
    if i < 49:
        funPar.append(np.cos(Neue[i])*np.sin(Neue[i])**(-8))
    else:
        funPar.append(np.sin(Neue[i])*np.cos(Neue[i])*(1/2))
plt.plot(x, funPar)
```

```python
plt.plot(x, SinCar, '.-', linewidth = 0.5, label = 'Seno Cardinal')
plt.plot(x, Neue, '*-', linewidth = 0.5, label = 'Funcion Multiplicada')
plt.plot(x, funPar,'-o', linewidth = 0.5, label = 'Funcion parametrica')
plt.plot(x, y,'-+', linewidth = 0.5,label = 'Funcion de ayer')
plt.grid()
plt.title('Comparacin de funciones')
plt.legend(loc = 'center left', bbox_to_anchor = (1, 0.5))
plt.xlabel('x')
plt.ylabel('y')
# Delimitar los ejes
plt.xlim(60, 80)
plt.ylim(-1, 1)
plt.savefig('Varias_Funciones', bbox_inches = 'tight')
plt.show()
```
